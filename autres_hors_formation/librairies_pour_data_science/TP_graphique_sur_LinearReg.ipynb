# TP numpy, matplotlib et seaborn : 

```python
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

import sklearn.linear_model
data_path = "~/Documents/openclassroom/data/"
```

```python
df = pd.read_csv(data_path+"autre_formation/"+"hubble_data.csv",
                dtype = float)
df.head(2)
```

```python
plt.scatter(df.distance, df.recession_velocity, marker="x")
plt.xlabel("distance")
plt.ylabel("recession velocity")
plt.title("Hubble data")
plt.show()
```

```python
## LINEAR REGRESSION : 
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
y = df.recession_velocity.values
X = df.distance.values.reshape(-1, 1)
lm.fit(y = y, X = X)
```

```python
## PLOT WITH MATPLOTLIB : 
plt.scatter(X, y, marker = "x")
plt.plot(X,[lm.coef_ * i + lm.intercept_ for i in X], 
         label = "linear regression", c="red")
plt.legend(loc="lower right")
plt.xlabel("X = distance")
plt.ylabel("y = recession velocity")
plt.title("Linear Regression on Hubble data")
plt.show()

```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
ax = ax.set(xlabel = "X = distance",
            ylabel = "y = recession velocity", 
            title = "Linear Regression on Hubble data")

plt.scatter(X, y, marker = "x", label = "data")
plt.plot(X,[lm.coef_ * i + lm.intercept_ for i in X], 
         label = "linear regression", c='red')
```
